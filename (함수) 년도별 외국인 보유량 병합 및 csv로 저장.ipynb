```python
import pandas as pd 

def hdays_year(year):
    # 휴장일 엑셀 파일 로드
    # 파일 경로 설정
    file_path = './excel/' + year + '_hdays.xls'
    hdays = pd.read_excel(file_path)
    # 불필요 컬럼 제거
    hdays.drop(columns=["요일구분", "비고"], inplace=True)
    # 불필요 문자 제거
    hdays["일자 및 요일"] = hdays["일자 및 요일"].str.replace("-","")
    # int로 데이터타입 변환
    hdays["일자 및 요일"] = hdays["일자 및 요일"].astype(int)
    # 휴장일 리스트 생성
    hdays_list = hdays["일자 및 요일"].values.tolist()
    
    # 1년치 영업일 리스트 생성
    bdays = pd.date_range(year+'-01-01', year+'-12-31', freq='B')
    # 빈 데이터프레임 생성
    df = pd.DataFrame()
    # bdays 컬럼 생성
    df['bdays'] = bdays
    df['bdays'] = df['bdays'].dt.strftime('%Y%m%d')
    df['bdays'] = df['bdays'].astype(int)
    
    bdays_list = df['bdays'].tolist()
    
    # 영업일에서 휴장일을 뺀 개장일
    odays = [x for x in bdays_list if x not in hdays_list]
    
    return odays
```

```python
hdays_year("2022")
```

```python
high50_df = pd.read_csv('./csv/2022_foreigner/data_1902_20220713.csv', encoding='CP949')
high50_df = high50_df.sort_values(by='외국인 지분율', ascending=False)[:50]
high50_list = high50_df["종목명"].tolist()
```

```python
import pandas as pd
import os

# 년도와 종목명을 입력하면 해당년도치 외국인 보유량 csv 병합
def for_to_csv():
    # 년도별 외국인 보유량 폴더 디렉토리 설정
    path = "./csv/foreigner/"

    file_list = []
    # 디렉토리 내 모든 csv 파일 로드 및 리스트에 추가
    for filename in os.listdir(path):
        if filename.endswith('.csv'):
            file_list.append(filename)
    # 순서대로 정렬
    file_list.sort()
    full_df = pd.DataFrame()
    
    # csv 파일 내 해당 년도와 해당 종목만 병합해서 csv 파일로 저장
    for file in file_list:
        temp_df = pd.read_csv(path+file, encoding='CP949')
#         temp_df["날짜"] = 
        # 불필요 컬럼제거
        temp_df = temp_df.drop(['종목코드', '종가', '대비', '상장주식수'], axis=1)
        temp_df = temp_df[temp_df['종목명'].isin(high50_df['종목명'].values)]
        temp_df[temp_df['종목명'].isin(high50_df['종목명'].values)]
        #         temp_df = temp_df[temp_df['종목명']==name]
        full_df = pd.concat([full_df, temp_df])
    
#     full_df = full_df.sort_values(by='외국인 지분율', ascending=False)
    full_df.reset_index(drop=True, inplace=True)
#     full_df = full_df.groupby('종목명')
    # csv 파일로 저장
    full_df.to_csv('./csv/외국인_상위50.csv', encoding='CP949')
#     full_df.to_csv('./csv/'+year+'_외국인_상위50기업.csv', encoding='CP949')
#     group = full_df.groupby(full_df['종목명']).apply(lambda full_df:full_df.sort_values(by=['외국인 지분율']))
#     full_df.to_csv('./csv/'+year+'_외국인_상위50기업.csv', encoding='CP949')
```
