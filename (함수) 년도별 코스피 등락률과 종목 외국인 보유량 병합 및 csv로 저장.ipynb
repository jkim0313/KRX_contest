```python
import pandas as pd
import numpy as np

# 외국인 보유량과 코스피 등락률 병합 함수
def for_read_csv():
    # 해당년도 외국인 보유량 csv파일 로드
    df = pd.read_csv('./csv/외국인_상위50.csv', encoding='CP949')
#     df = pd.read_csv('./csv/'+year+'_외국인_상위50기업.csv', encoding='CP949')
    # 불필요 컬럼 제거
#     df.drop(['Unnamed: 0'], axis=1, inplace=True)
    
    # 필요한 컬럼만 남긴다
    df = df[['종목명', '외국인 지분율', '외국인 한도소진율', '등락률']]
    # "등락률" 컬럼명을 "기업 등락률"로 변겅
    df.rename(columns = {'등락률' : '기업 등락률'}, inplace=True)
    
    # 해당년도 코스피 csv 파일 로드
    kospi = pd.read_csv('./csv/kospi_4yrs.csv', encoding='CP949')
    
    kospi = kospi.sort_values(by='일자', ascending=True)
    kospi.reset_index(drop=True, inplace=True)
    
    df["코스피"] = kospi["종가"]
    df["코스피 등락률"] = kospi["등락률"]
    
    # 날짜 컬럼 추가
    df["날짜"] = kospi["일자"]
    # 날짜를 인덱스로 설정
#     df = df.set_index('날짜')
    
    # 코스피 등락 평균
    df["코스피 등락평균"] = df["코스피 등락률"].mean()
    # 기업 등락 평균
    df["기업 등락평균"] = df["기업 등락률"].mean()
    
    # 변동성 계산 식
    df["변동성"] = (df['코스피 등락률'] * df['기업 등락률']) > 0 & (abs(df['코스피 등락률']) - abs(df['기업 등락률'])<0)
    # 변동성 계산 식 적용
    df["변동성"] = df['변동성'].apply(lambda x : 1 if x==True else 0 )
    # 변동성 수치 계산
    df["변동성 수치"] = abs(df['코스피 등락평균'] - df['코스피 등락률']) - abs(df['기업 등락평균'] - df['기업 등락률'])
    
    
    com_name = df["종목명"]
    
    for n in com_name:
        
        # 공분산
        kospi_cov = np.cov(df["코스피 등락률"], df["기업 등락률"])[1,0]
        # 분산
        kospi_var = np.var(df["코스피 등락률"])

        beta = kospi_cov / kospi_var

        print(n, ": ", beta)

    # csv로 저장
    df.to_csv('./csv/_코스피등락률.csv', encoding='CP949')
#     df.to_csv('./csv/'+year+'_외국인_상위50기업'+'_베타.csv', encoding='CP949')
```

```python
for_read_csv()
```
