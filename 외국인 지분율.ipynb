```python
foreign_trade_200102 = pd.read_csv('./csv/유가증권/투자자별 거래실적(1분)/2020/[유가증권]투자자별 거래실적(1분)(주문번호-1300-45)_20200102.csv', encoding='CP949')

foreign_trade_200102_list = foreign_trade_200102["종목명"].tolist()

# 제거해야 할 종목 리스트
drop_list = [x for x in foreign_trade_200102_list if x not in comp_name_list]        
drop_list

# 불필요 컬럼 제거
foreign_trade_200102.drop(["시장ID"], axis=1, inplace=True)
# 필요한 종목만 남긴다
foreign_trade_200102.drop(foreign_trade_200102[foreign_trade_200102["종목명"].isin(drop_list)].index, inplace=True)
# 외국인 코드만 남긴다
foreign_trade_200102 = foreign_trade_200102[foreign_trade_200102["투자자구분코드"]==9000]
# 시간외 거래 제거
foreign_trade_200102.drop(foreign_trade_200102[foreign_trade_200102["거래시각"] < 900].index, inplace=True)
foreign_trade_200102.drop(foreign_trade_200102[foreign_trade_200102["거래시각"] > 1530].index, inplace=True)
foreign_trade_200102.reset_index(drop=True, inplace=True)

foreign_trade_200102.head(2)
```
